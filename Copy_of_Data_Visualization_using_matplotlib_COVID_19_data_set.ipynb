```python
import csv 
import numpy as np
with open('covid_19_data.csv','r') as file: #Using context management to handle opening of file. Opened in read mode 
    csv_reader=csv.DictReader(file) #Returns a dictionary that can be iterated upon, each row an ordered dictionary
    sum_of_cases=np.array([0, 0, 0])#initializes a list that contains 3 items
    for row in csv_reader:
        sum_of_cases[0] +=float(row['Confirmed'])#converts dictvalue to string and increaments item in list
        sum_of_cases[1] +=float(row['Deaths'])
        sum_of_cases[2] +=float(row['Recovered'])
print(sum_of_cases)
```

```python
cases = ['Confirmed', 'Deaths', 'Recovered']
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(cases,sum_of_cases)
plt.title('Recorded Cases', size=14)
```

```python
plt.pie(sum_of_cases,labels=cases,explode=(0,0.4,0)) #explode=(0,0.4,0) extracts the second item in the list
plt.title('A pie chart of Cases Recorded', size=15)
```

## Rate of Cases

This seeks to visualize the number of cases recorded on each day and location to bring to light the trend in the data.

```python
with open('covid_19_data.csv','r') as file:
    csv_reader=csv.DictReader(file)
    reported_cases={'Confirmed':[],'Deaths':[],'Recovered':[]}#Initialization of a dictionary of lists of reported cases
    observed_date=[]
    for row in csv_reader:
        reported_cases['Confirmed'].append(float(row['Confirmed']))#appends values on each row to the list called here
        reported_cases["Deaths"].append(float(row['Deaths']))
        reported_cases['Recovered'].append(float(row['Recovered']))
        observed_date.append(row['ObservationDate'])#list of observation dates
        
```

```python
fig = plt.figure(figsize=(10,6))
plt.plot(observed_date,reported_cases['Confirmed'])
plt.xticks(rotation='vertical', size=7) #This changes the direction and size of the ticks on the axis
plt.xlabel('observed date', size=12)
plt.ylabel('Confirmed', size=12)
plt.title('Confirmed Cases', size=20)
```

```python
fig = plt.figure(figsize=(10,4))
plt.bar(observed_date,reported_cases['Confirmed'])
plt.xticks(rotation='vertical', size=7)
plt.xlabel('observed date', size=12)
plt.ylabel('Confirmed', size=12)
plt.title('Barchart of Confirmed cases', size=18)
```

```python
fig=plt.figure(figsize=(10,6))
plt.plot(observed_date,reported_cases['Deaths'],)
plt.xticks(rotation='vertical', size=7)
plt.xlabel('observed date', size=12)
plt.ylabel('Deaths', size=12)
plt.title('Death Cases', size=20)
```

```python
fig=plt.figure(figsize=(10,6))
plt.plot(observed_date,reported_cases['Recovered'],)
plt.xticks(rotation='vertical', size=7)
plt.xlabel('observed date', size=12)
plt.ylabel('Recovered', size=12)
plt.title('Recovered Cases', size=20)
```

```python
fig = plt.figure(figsize=(10,6))
plt.bar(observed_date,reported_cases['Recovered'], label='Recovered',)
plt.bar(observed_date,reported_cases['Deaths'], label='Deaths',)
plt.xticks(rotation='vertical', size=7)
plt.xlabel('observed date', size=12)
plt.ylabel('Confirmed', size=12)
plt.title('Comparing the rate of death and recovery', size=18)
plt.legend()
```
